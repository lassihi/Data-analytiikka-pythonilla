## Aineiston suodatukset

```python
# poimitaan työntekijät, joiden palkka on alle 2000 euroa:
df[df["palkka"] > 2000]
```

```python
# työntekijät, joiden palkka on korkeintaan 2000 euroa ja ikä vähintään 50 vuotta

df[(df["palkka"] <= 2000) & (df["ikä"] >= 50)]
```

```python
# työntekijät, joiden palkka on korkeintaan 2000 euroa tai ikä vähintään 50 vuotta

df[(df["palkka"] <= 2000) | (df["ikä"] >= 50)]
```

## Aineiston muutokset

```python
# muodostetaan uusi muuttuja "syntymävuosi"
df["syntymävuosi"] = 2026 - df["ikä"]
df.head()
```

```python
# luodaan muuttuja, joka laskee kuinka montaa etuisuutta työntekijä on käyttänyt:
df["käyttö"] = df[["työterv", "lomaosa", "kuntosa", "hieroja"]].count(axis=1)
```

```python
df.head()
```

```python
# muokataan rivillä 4 olevan henkilön ikää
df.loc[4, "ikä"] = 124
df.head()
```

```python
# tiputetaan pois ne rivit, joilla on puuttuvia arvoja työterv- tai lomaosa-muuttujissa:
df.dropna(subset=["työterv", "lomaosa"])
```

```python
# järjestetään työntekijät iän mukaan vanhimmasta nuorimpaan
df.sort_values(by="ikä", ascending=False)
```

```python
# tiputetaan pois muuttujat ikä ja palkka:
df.drop(columns=["ikä", "palkka"])
```

```python
# tiputetaan rivit 1 ja 2:
df.drop(index=[1, 2])
```
